## Scatter

```python
x=np.arange(1,6)
y=np.arange(10,15)
print(x)
print()
print(y)
print('________visualization______')
plt.scatter(x,y)
plt.show()
```

```python
x=np.arange(1,6)
y=np.arange(10,15)
print(x)
print(y)
plt.scatter(x,y,color='r',marker='+')

plt.show()
```

```python
x=np.arange(1,25,3)
y=[9,6,7,4,65,3,10,11]
print(x)
print(y)
plt.scatter(x,y,color='g',marker='+')
plt.show()
```

```python
## plot
```

```python
x=np.arange(1,25,3)
y=[9,6,7,4,65,3,10,11]
print(x)
print(y)
plt.title("Mornig Batch")
plt.xlabel('values of x')
plt.ylabel('Values of y')
plt.plot(x,y,color='g', marker='+')
plt.show()
```

## Legend

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r')
plt.plot(co,b,color='g')
plt.plot(co,c,color='b')
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='g',label='co and b')
plt.plot(co,c,color='b',label='co and c')
plt.legend(loc=0)
plt.show()
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='b',label='co and b')
plt.plot(co,c,color='g',label='co and c')
plt.legend(loc=0,shadow=True,fontsize='small')
plt.show()
```

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]
plt.figure(figsize=(10,5))
plt.plot(co,a,color='r',label='co and a')
plt.plot(co,b,color='b',label='co and b')
plt.plot(co,c,color='c',label='co and c')
plt.legend(loc=0,shadow=True,fontsize='x-large')
plt.show()
```

## Sub plot

```python
co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
plt.subplot(2,1,1)
plt.plot(co,a,color='r',label='co and a')
plt.subplot(2,1,2)
plt.plot(co,b,color='g',label='co and b')
plt.legend(loc=0)
```

## Pie chart

```python
x=[40,30,20,10,1]
y=['java','python','c#','PHP','c++']
plt.figure(figsize=(7,7))
plt.pie(x,labels=y)
```

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','c++']
plt.figure(figsize=(7,7))
plt.pie(x,labels=y,shadow=True)
```

```python
x=[40,30,20,10,1]
y=['java','python','C#','PHP','C++']
plt.figure(figsize=(7,7))
plt.pie(x,labels=y,shadow=True,autopct='%1.1f%%')
```

```python
x = [40,30,20,10,1]
y = ['java','Python','c#','PHP','c++']
plt.figure(figsize = (7,7))
plt.pie(x , labels = y , shadow = True , autopct = '%1.1f%%' , explode = [0,0.4,0,0.2,0])
```
